```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#sample Air Quality data for Indian cities
data = pd.read_excel('Downloads/air_quality_data (3).xlsx')
air_quality_df = pd.DataFrame(data)
data
```

```python
print(air_quality_df.head())
print(air_quality_df.info())
```

```python
# Convert the column to numeric, forcing errors to NaN
air_quality_df['PM2.5 Annual Average'] = pd.to_numeric(air_quality_df['PM2.5 Annual Average'], errors='coerce')

# Now calculate the mean, ignoring NaN values
average_pm25 = air_quality_df['PM2.5 Annual Average'].mean()

# Print the result correctly
print("\nAverage PM2.5 across cities:", average_pm25)
```

```python
# Convert 'NO2 Annual Average' to numeric, forcing errors to NaN
air_quality_df['NO2 Annual Average'] = pd.to_numeric(air_quality_df['NO2 Annual Average'], errors='coerce')

# Now find the city with the highest NO2 levels
city_highest_no2 = air_quality_df.loc[air_quality_df['NO2 Annual Average'].idxmax(), 'City/Town']

# Print the result correctly (fixing the typo: "/n" → "\n")
print("\nCity with highest NO2 levels:", city_highest_no2)
```

```python
correlation_pm25_no2=air_quality_df['PM2.5 Annual Average'].corr(air_quality_df['NO2 Annual Average'])
print("\n correlation between PM2.5 and NO2:", correlation_pm25_no2)
```

```python
#bar chart of PM2.5 levels by city
plt.figure(figsize=(50,25))
sns.barplot(x='City/Town',y='PM2.5 Annual Average',data=air_quality_df)
plt.title('PM2.5 levels by city')
plt.xlabel('city')
plt.ylabel('PM2.5')
plt.xticks(rotation=45)
plt.show()
```

```python
#scatter plot of PM2.5 vs NO2
plt.figure(figsize=(8,6))
sns.scatterplot(x='PM2.5 Annual Average',y='NO2 Annual Average',data=air_quality_df)
plt.title('PM2.5 vs NO2 level')
plt.xlabel('PM2.5')
plt.ylabel('NO2')
plt.show()
```
